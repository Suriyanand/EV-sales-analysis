```python
import seaborn as sns
import matplotlib.pyplot as plt
import mplcursors
```

```python
sns.catplot(x = 'percent_change',y = 'values',hue = 'name',data = df7,kind = 'point')
```

```python
bitcoin_data = df7[df7['name'] == 'Bitcoin']




# Plotting with Seaborn
sns.set_style("whitegrid")
ax = sns.lineplot(x='percent_change', y='values', data=bitcoin_data, marker='o')

def on_plot_hover(event):
    if event.inaxes == ax:
        x_data = ax.lines[0].get_xdata()
        y_data = ax.lines[0].get_ydata()
        index = int(round(event.xdata))
        if 0 <= index < len(x_data):
            plt.text(x_data[index], y_data[index], 
                     f'Percent Change: {x_data[index]:.2f}, USD Price: {y_data[index]:.2f}',
                     ha='right', va='bottom', fontsize=8, color='black', bbox=dict(facecolor='white', alpha=0.7))
            plt.draw()

plt.gcf().canvas.mpl_connect('motion_notify_event', on_plot_hover)

plt.title('Bitcoin USD Price Over Time')
plt.xlabel('Percent Change')
plt.ylabel('USD Price')

plt.show()
```
